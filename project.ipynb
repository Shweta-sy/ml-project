```python
# LINEAR REGRESSION
#REGRESSION OF GLUCOSE WITH BLOODPRESSURE

glucose=df['Glucose']

bloodpressure=df['BloodPressure']

x=glucose
y=bloodpressure

slope,intercept,r,p,std_err=stats.linregress(x,y)

def myfunc(x):
  return slope*x+intercept

mymodel=list(map(myfunc,x))

plt.scatter(x,y,c='c')

plt.plot(x,mymodel,c='k')

plt.show()



```

```python
#printing the relationship regarding the linear regression
# r should be between -1 to 1 




print(r)

```

```python
#applying concept of polynomial regression on 2 factors glucose and bloodpressure of given data set



import numpy
import matplotlib.pyplot as plt

glucose=df['Glucose']

bloodpressure=df['BloodPressure']

x=glucose
y=bloodpressure



mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

myline = numpy.linspace(1, 22, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()





#printiting r_squared value to know how well does data fitin a polynomial regression

print(r2_score(y, mymodel(x)))
```

```python
#applying multiple regresssion on the given data set to predict Bloodpressure of a person with 
#glucose level=100
#insulin level=100
#age=80





df=pd.read_csv("/content/diabetes.csv")

x=df[['Glucose','Insulin','Age']]
y=df[['BloodPressure']]
regr=linear_model.LinearRegression()
regr.fit(x,y)
predictedBloodpressure=regr.predict([[100,100,80]])
print(predictedBloodpressure)
print(df)
```
